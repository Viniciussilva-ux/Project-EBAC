```python
import pandas as pd
# O nome do arquivo carregado é 'MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv'
df = pd.read_csv('MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv')
df.head(10)
```

Os campos do nosso dataframe são:

**Title:** Nome do produto.


**Marca:** A marca do produto.


**Preco_Normal:** O preço em que o produto costuma ser vendido quando não há desconto.


**Preco_Desconto:** O preço vendido após o desconto ser aplicado.


**Preco_Anterior:** Preço em que era comercializado o produto antes do desconto aplicado.


**Desconto:** Total de desconto aplicado.






As colunas que aparecem com valores 0 são para os produtos onde não tivemos descontos aplicados.


As categorias estão em espanhol!

# 1 - Traga a média e a mediana dos preços - coluna Preco_Normal - por categoria de produto.
# Identifique as categorias que parecem ter um valor de média abaixo ou acima da mediana.

```python
df.groupby('Categoria')['Preco_Normal'].mean().reset_index().sort_values(by='Categoria', ascending=False)
```

```python
df.groupby('Categoria')['Preco_Normal'].median().reset_index().sort_values(by='Categoria', ascending=False)
```

Digite aqui as categorias: Ao que parece, a categoria lacteos tem uma média muito acima da mediana e em seguida congelados e frutas.

# 2 - Traga o desvio padrão por categoria de produto.
# Qual o comportamento da média e mediana nas categorias com maior desvio?

```python
df.groupby('Categoria')['Preco_Normal'].std().reset_index().sort_values(by='Preco_Normal', ascending=False)
```

Na categoria lacteos a média tem o valor muito acima da mediana e na categoria belleza-y-cuidado-personal a mediana ficou relativamente próxima da média.

# 3 - Plot um boxplot da distribuição do Preco_Normal para a categoria que você identificou que tem o maior desvio padrão. Como é a distribuição desses dados segundo o boxplot? Você identifica muitos outliers?

Segundo o Boxplot a distribuição dos dados não apresenta outliers.

```python
fig=px.box(df.loc[df['Categoria'] == 'lateos'], y= 'Preco_Normal', title='BoxPlot de Preco_Normal para lacteos')
fig.show()
```

# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
media_desconto = df.groupby('Categoria')['Desconto'].mean().reset_index().sort_values(by='Desconto', ascending=False)
fig = px.bar(media_desconto, x='Categoria', y='Desconto', title='Média de Desconto por Categoria')
fig.show()
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
media_desconto_categoria_marca = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index().sort_values(by='Desconto', ascending=False)
fig = px.treemap(media_desconto_categoria_marca,
                   path=['Categoria', 'Marca'],
                  values='Desconto',
                  title='Média de Desconto por Categoria e Marca')
fig.show()
```
